# MIMOSA2 data comparison

In this notebook we exploit STELLA to obtain the OTUs-metabolite for the data used in the MIMOSA2 workframe by [Noecker *et al.* (2021)](https://www.biorxiv.org/content/10.1101/2021.09.14.459910v1).
The code is available at [https://borenstein-lab.github.io/MIMOSA2shiny/](https://borenstein-lab.github.io/MIMOSA2shiny/).

Once again, the first step is to create a pathway-OTUs matrix.
Then this matrix is multiplied by the score matrix obtained from the MetaCyc table with the list of interesting metabolic pathways and metabolites (available online at [https://metacyc.org/group?id=biocyc17-60476-3862372913](https://metacyc.org/group?id=biocyc17-60476-3862372913)), from which we obtain the patient-metabolite matrix.

MIMOSA2 data is located at `../data/MIMOSA2/`, and there is a *Readme* file giving information about how to move inside the directory.

We report some guidelines provided by Noecker *et al.* in their MIMOSA2 reference paper/framework:

> Specifically, microbiome data can be provided in the form of either a taxa-based table of 16S rRNA microbiome data processed in one of several ways, or a function-based table of KEGG Ortholog abundances annotated from metagenomic sequencing data.\
[...]\
Input data files should be tab-delimited or comma-delimited tables with a row for each feature and a column for each sample, with the first row being a header of names for each column. There should not be duplicate features or samples. If sample IDs differ between the microbiome and metabolome data files, MIMOSA will only analyze samples that are present in both.

The problem is that in the MIMOSA2 database there is no file with an "ID" column to identify the pathway belonging (e.g. "PWY-7001")... how to fix this?
We need the ID in order to link properly the OTUs to the metabolites consumed/produced, since `1-pathway-otus.ipynb` yields a patient-pathway matrix, and `2-metacyc-score.ipynb` gets a pathway-metabolite matrix.
By multiplying these matrices one gets the patient-metabolites matrix starting from a set of OTUs for each patient.

There are however some files in the MIMOSA2 database that have a label `RxnID` to identify the reaction to which some reagent belong. 
One may exploit this specific label to link them to the OTUs of a patient.

MIMOSA2 exploits [Franzosa *et al. (2018)*](https://www.nature.com/articles/s41564-018-0306-4) data to analyze human gut microbiota. Data is available at [https://www.metabolomicsworkbench.org/data/DRCCMetadata.php?Mode=Project&ProjectID=PR000677](https://www.metabolomicsworkbench.org/data/DRCCMetadata.php?Mode=Project&ProjectID=PR000677).
It is a pretty big database, of size 106 GB.

Some useful info about the data contained in the MIMOSA2 dataset follow.

## `general` directory

As reported in the article,

- `FaithMedia*`: Nutrient media contents used for the ten-species simulations (`FaithMedia_AGORA_F_final.csv`), and the HMP-based simulations (`FaithMedia_AGORA_expanded_revised.csv`);
- `Dictionary*`: Compound dictionary file used for the ten-species simulations (`Dictionary_AGORA_complete.csv`) and the HMP-based simulations (`Dictionary_AGORA_all_1.csv`);
- `hmdb_brite_metabolite_classes.txt`: Metabolite classification system for all metabolites in the 10-species datasets.

```python
dfDictionaryAGORA = pd.read_csv("../data/MIMOSA2/general/Dictionary_AGORA_complete.csv")

# Rearrange columns
dfDictionaryAGORA = dfDictionaryAGORA[["medium"] + [col for col in dfDictionaryAGORA.columns if col != "medium" ]]
dfDictionaryAGORA = dfDictionaryAGORA[["Primary"] + [col for col in dfDictionaryAGORA.columns if col != "Primary"]]
#dfDictionaryAGORA.fillna("", inplace=True)

# Create dictionary: code -> species
dictCodeToSpecies = pd.Series(dfDictionaryAGORA.Primary.values, index=dfDictionaryAGORA.medium).to_dict()
dfDictionaryAGORA
```

## `mainDataset` directory

- `allSpeciesMain.txt`: Species abundances across all time points and all simulation runs --- **NB** There are only 10 different species;
- `allMetabolitesMain.txt`: Metabolite concentrations across all time points and simulation runs

```python
MIMOSAReactionMetabolites = pd.read_csv("../data/MIMOSA2/mainDataset/MIMOSA_reactionsMetabolites.txt", sep="\t")
MIMOSAReactionMetabolites.set_index("Compound", inplace=True)
# Sort columns (alphabetically) as next indexes
MIMOSAReactionMetabolites = MIMOSAReactionMetabolites.reindex(sorted(MIMOSAReactionMetabolites.columns), axis=1)
MIMOSAReactionMetabolites
```

```python
MIMOSASpeciesReactions = pd.read_csv("../data/MIMOSA2/mainDataset/MIMOSA_speciesReactions.txt", sep="\t")

# RxnID = "Reaction ID"
MIMOSASpeciesReactions.set_index("RxnID", inplace=True)
# Sort indexes (alphabetically) as previous columns
MIMOSASpeciesReactions.sort_index(inplace=True)
MIMOSASpeciesReactions
```

### Dataframes intersection

The `RxnID` index (meaning "Reaction ID") can be intersected between the *Reaction-Metabolite* dataframe (where $M_{ij} \in \{-1, 0, 1\}$, according to deficency/balance/exceeding of the specific compound I think) and the *Species-Reaction* one (binary matrix) in order to link the compound to the specific species.
In this way a *Metabolite-Species* dataframe is obtained by matrix product.

```python
# Matrix product (intersection) of MIMOSA dataframes
prod = MIMOSAReactionMetabolites @ MIMOSASpeciesReactions 
prod2 = prod.copy()
prod2.index = prod2.index.map(dictCodeToSpecies) # NaNs for some indexes
prod2 = prod2[prod2.index.notnull()]
prod2
```

```python
# Figure parameters
ax = plt.subplots(figsize=(12, 16))
plt.title("Compound-species heatmap")

# Plot
sns.heatmap(prod2, vmin=-5, vmax=5, cmap="seismic")
```

```python
# Figure parameters
ax = plt.subplots(figsize=(14, 6))
plt.ylim(-4, 4)
plt.title("Species value distributions")
sns.set_theme(style="whitegrid")

# Plot
sns.violinplot(data=prod2, palette="viridis_r")
sns.stripplot(data=prod2, linewidth=1, size=3, color=".3")

# Tweak the visual presentation
sns.despine(trim=True, left=True)
```
