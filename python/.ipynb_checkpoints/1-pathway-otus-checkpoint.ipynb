# Pathway-OTU Matrix

The goal of this notebook is to obtain a patient-pathway matrix containing an estimate of the total presence of a specific pathway in each host.
This estimate is calculated as the sum of the presence of the pathway over all the taxonomies in the host itself:
$$
  P_{ij} = \sum_{k=1}^{n_i} a_{ik} d_{kj} ,
$$
where $n_i$ is the number of bacteria in the $i$-th host, $a_{ik} \in \mathbb{R}$ is the abundancy of bacteria $k$ in the host $i$ and $d_{kj} = 1$ if the $j$-th pathway is present in the taxonomy $k$ ($0$ otherwise).

When using single-patient data, this score is an abundance or percentage of abundance in the subject, patient or control.

## MACADAM database
Information on the metabolic pathways present in a given taxonomy were obtained from the metabolic pathways database for microbial taxonomic groups (MACADAM).

Note on the OTUs ID from the original paper https://academic.oup.com/database/article/doi/10.1093/database/baz049/5480179:

>To link this information, we connected data from PGDBs, the FAPROTAX and the IJSEM phenotypic database with
>their NCBI taxonomy ID (taxID).
>This taxID is a unique time-stable identifier because of its interoperability with
>other taxonomic databases.
>Briefly, in MACADAM, each organism taxonomy has seven taxonomic ranks: *superkingdom, phylum, class, order, family, genus and species.*
>Each rank and organism is described by its NCBI `taxID`.
>Each MACADAM organism is therefore associated with its numeric lineage formed by seven taxID.
>For example, *E. coli* is described as *Bacteria, Proteobacteria, Gammaproteobacteria, Enterobacterales, Enterobacteriaceae, and Escherichia*.
>E. coli is linked to this lineage `2.1224.1236.91347.543.561.562` in MACADAM.
>
>However, even if MACADAM data are not associated with minor ranks, e.g. subclasses or subgenera, these minor ranks are kept in MACADAM so that users can find functional information at these minor ranks.

## Autism Spectrum Disorder [(ASD)](https://www.wikiwand.com/en/Autism_spectrum)
In this section we obtain the patwhay-host matrix for patients affected by ASD. 

### Dataframe creation and formatting

```python
# Read MACADAM querys
filesToRead = glob("../data/MACADAM/Results/*.tsv")
pathwaysMACADAM = pd.concat((pd.read_csv(f, sep="\t") for f in filesToRead), ignore_index=True)

# Fix HTML tags to UNICODE text
pathwaysMACADAM["Metabolic Pathway"] = pathwaysMACADAM["Metabolic Pathway"].apply(lambda x: html.unescape(x))

# Filter database
pathwaysMACADAM = pathwaysMACADAM[["Metabolic Pathway", "Target taxonomies", "MetaCyc URL"]]
```

Note that the following data is misformatted when downloaded from MACADAM.
Some entries of the "Complete" are made of only 1 word.
Thus, we shall thus get rid of them.

In the following we are selecting only well-formatted data and choose from it the correct bacteria names
to create a new dataframe with this new data.

```python
listOTUs = pd.read_csv("../data/MACADAM/ListOTUs/Genus_Fam_Ord_Phylum_complete_name_Autism.txt",
                       sep=";", names=["Short", "Complete"])
listOTUs.reset_index(drop=True, inplace=True)
listOTUs.replace({r"\s+": " ", r"\s?c__\s*": "c__"}, regex=True, inplace=True)
```

```python
# Select OTUs missing in MACADAM queries
missingOTUs = listOTUs[~(listOTUs["Short"].isin(pathwaysMACADAM["Target taxonomies"]))]["Complete"]
subst = missingOTUs.copy()
# Take the last 2 "words"
subst = subst.apply(lambda x: x.split()[-2:])
```

```python
# Some weird string manipulations to obtain the correct bacteria names at the correct filus order
subst2 = subst.to_frame(name="Short")
subst2["Complete"] = missingOTUs
subst2["Backup"] = subst2["Short"]
subst2["Short"] = subst2["Short"].str.join(" ")
subst2 = subst2[~subst2["Short"].str.contains("_")]
subst2.drop(columns="Short", inplace=True)
subst2.rename(columns={"Backup": "Short"}, inplace=True)
subst2 = pd.DataFrame(subst2["Short"].tolist(), columns=["b1", "b2"])
```

```python
listOTUs["Short Correlated"] = listOTUs["Short"]
listOTUs["Short Correlated"] = listOTUs["Short"].map(subst2.set_index("b2")["b1"])

listOTUs["Short Correlated"].fillna(listOTUs["Short"], inplace=True)
substitutions = {r"\s+": " ", r"\.\.": ".",
                 r"\sp__\s*": "p__",
                 r"\sc__\s*": "c__",
                 r"\so__\s*": "o__",
                 r"\s__\.": "__",
                }
listOTUs["Complete"].replace(substitutions, regex=True, inplace=True) # fix spaces
```

```python
# Filter pathways
pathwaysMACADAM = pathwaysMACADAM[pathwaysMACADAM["Target taxonomies"].isin(listOTUs["Short Correlated"])]
pathwaysMACADAM["Val"] = 1
pathwaysMACADAM.drop(["Target taxonomies"], axis=1)

df = pathwaysMACADAM.pivot(index="MetaCyc URL", columns="Target taxonomies", values="Val").fillna(0)
df.to_csv("./tmp/asd/matrix-pathway-OTU.csv", index=False)
```

```python
ax = plt.subplots(figsize=(12, 8))

# Discrete (binary) colorbar
bw = ("white", "black")
cmap = matplotlib.colors.LinearSegmentedColormap.from_list("Custom", bw, len(bw))

# Plot dataframe
ax = sns.heatmap(df, cmap=cmap, linecolor="lightgray")

# Manually specify colorbar labelling after it has been generated
matplotlib.colors.colorbar = ax.collections[0].colorbar
matplotlib.colors.colorbar.set_ticks([0.25, 0.75])
matplotlib.colors.colorbar.set_ticklabels(["0", "1"])

df.describe()
```

```python
# Read abundancy files
filesToRead = glob("../data/MACADAM/Filebase/Autism/*.csv")
for f in filesToRead:
    abundancies = pd.read_csv(f, sep=",")
    
    # Make new indexes
    indexes = list(abundancies["disease_name"])
    abundancies.drop(["disease_name"], axis=1, inplace=True)
    
    # Add numerical label to indexes
    for i in range(len(indexes)):
        #indexes[i] += (" "+str(i+1))
        indexes[i] += " "+f"{(i+1):02d}"
    
    # Update dataframe indexes
    abundancies.index = indexes 

    # Adjust name entries
    substitutions = {r"\.\.": ".", r"\sp__": "p__", r"\sc__": "c__", r"\so__": "o__", r"__\.": "__"}
    abundancies.replace(substitutions, regex=True, inplace=True)
    
    # Take most contributing values and normalize them
    abundancies = np.power(10, abundancies.astype(float), dtype=float)
    abundancies[abundancies<np.power(10, -7., dtype=float)] = 0
    
    # Normalize, i.e. divide each row (axis=0) element by the sum of the corresponding column (axis=1)...
    abundancies = abundancies.div(abundancies.sum(axis=1), axis=0)
    # ...but watch out for computations with very small numbers
    abundancies[abundancies<np.power(10, -7., dtype=float)] = 0
    abundancies["Patient"] = abundancies.index
    vals = list(abundancies.columns)
    vals.remove("Patient")
    
    # Fix names
    dfLong = pd.melt(abundancies, id_vars="Patient", value_vars=vals, var_name="Name", value_name="Value")
    
    substitutions = {r"\s+": " ", r"\.\.": ".",
                     r"\sp__\s*": "p__",
                     r"\sc__\s*": "c__", "c__ ": "c__",
                     r"\so__\s*": "o__",
                     r"__\.": "__"}
    dfLong["Name"].replace(substitutions, regex=True, inplace=True)
    listOTUs["Complete"].replace({r"\s+": " "}, regex=True, inplace=True)
    renameColumns = dict(zip(listOTUs["Complete"], listOTUs["Short Correlated"]))
    dfLong["Name 2"] = dfLong["Name"].map(renameColumns)
    
    # The old name is now useless
    dfLong.drop(columns="Name", inplace=True)
    dfLong.rename(columns={"Name 2": "Name"}, inplace=True)
    
    # Do not want to miss the NaNs...
    dfLong2 = dfLong.groupby(by=["Patient", "Name"])["Value"].sum().reset_index()
    #dfLong2.drop_duplicates(inplace=True)
    
    dfLong2 = dfLong2.pivot(index="Patient", columns="Name", values="Value")#.fillna(0)
    
    xxx = [x for x in list(dfLong2.columns) if x in list(df.columns)]
    
    df2 = df[df.columns.intersection(xxx)]
    bindMe = df2.append(dfLong2[dfLong2.columns.intersection(xxx)])
    prod = bindMe @ bindMe.T
    
    columnsToKeep = prod.index[-len(dfLong2.index):]
    columnsToRemove = [i for i in prod.index if i not in columnsToKeep]
    indexesToKeep = df2.index
    indexesToRemove = [i for i in prod.index if i not in indexesToKeep]
    
    # Save output
    prod2 = prod.drop(index=indexesToRemove, columns=columnsToRemove)
    prod2.to_csv("./tmp/asd/Pathway_Host_"+f[32:], index=False)    
    prod2.set_index(indexesToKeep)
    prod2.to_csv("./tmp/asd/Pathway_Host_"+f[32:-4]+"_index.csv", index=True)    
```

Summing up, we obtained a patient-pathway matrix produced has as rows the pathways names and as columns the patients.
This matrix contains an estimate of the total presence of a specific pathway in each host.
This estimate is calculated as the sum of the presence of the pathway over all the taxonomies in the host itself.
