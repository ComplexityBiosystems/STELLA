# $F_1$-score

We want to compute here the $F_1$ score between STELLA-obtained data and experimental data from [Kang et al., 2017](https://www.sciencedirect.com/science/article/pii/S1075996417302305?via%3Dihub#appsec1), in order to have a benchmark value to which lean on.
This calculation is two-fold, since we obtain in this way both the validation of our dataset *and* the comparison between different methods for obtaining metabolites.
In fact, the comparison between different $F$-scores was already inspected and determined by [Yin et al., 2020](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC7746778/), resulting in a poor performance of all pipelines in identifying differential metabolites.

One may expect a similar result for STELLA, as shown in the previous picture.
There is no significant difference between healty and autistic patients, while we expected a two-block heatmap that marks the difference between these categories.


## The $F_\beta$ score

The more general F-score $F_{\beta }$, that uses a positive real factor $\beta$ , where $\beta$ is chosen such that _recall_ is considered $\beta$ times as important as _precision_, is computed as

$$
  F_\beta = \frac{(1+\beta^2) \mathrm{TP}}{(1+\beta^2)\mathrm{TP} + \beta^2 \mathrm{FN} + \mathrm{FP}},
$$

where TP stands for true positive, FN stands for false negative, FP stands for false positive and $\beta > 0$.

The $F_1$ score is the harmonic mean of the precision and recall.
The more generic $F_{\beta }$ score applies additional weights, valuing one of precision or recall more than the other.
The highest possible value of an F-score is 1.0, indicating perfect precision and recall, and the lowest possible value is 0, if either the precision or the recall is zero. 

Two commonly used values for $\beta$ are 2, which weighs recall higher than precision, and 0.5, which weighs recall lower than precision. 

Recall the relationship between true/false positive/negative values in the following picture, adapted from [Wikipedia](https://en.wikipedia.org/wiki/File:PPV,_NPV,_Sensitivity_and_Specificity.svg#/media/File:PPV,_NPV,_Sensitivity_and_Specificity.svg):


![](pix/sensitivityspecificity.svg "")

We shall then label STELLA output values as follows:

- "TP" if $m$ is labeled as ASD and $m$'s value lies the closest to the benchmark ASD value compared to the other, and "FN" otherwise.
- "TN" if $m$ is labeled as neurotypical and $m$'s value lies the closest to the benchmark neurotypical value compared to the other, and "FP" otherwise

```python
def Fb(tp, fn, fp, beta=1):
    """Compute F_beta score for a given number of true/false positive/negative values
    Parameters
    ----------
    tp: number of true positive values
    fn: number of false negative values
    fp: number of false positive values
    beta: value for which recall is considered beta times as important as precision:
          Default value is beta=1
    """
    return ((1+beta*beta)*tp) / ((1+beta*beta)*tp + beta*beta*fn + fp)
```

```python
# Import whole STELLA data
dfSTELLA = pd.read_csv("asd/patient-metabolite-matrix.csv")
dfSTELLA.set_index("Unnamed: 0", inplace=True)

# NB negative metabolites means that these compounds are consumed among all the selected pathways.
# We thus keep only positive (i.e, produced) metabolites in order to compare them with the experimentally found ones.
dfSTELLA = dfSTELLA[dfSTELLA[dfSTELLA.columns] > 0]
dfSTELLA.dropna(inplace=True)
dfSTELLA
```

```python
# Read Kang Excel file (supplementary data of the article)
# NOTE: NMR metabolites (Unit: micromole per g (dry stool))
dfKang2017 = pd.read_excel("../data/kang2017/1-s2.0-S1075996417302305-mmc2.xlsx", "metabolites")
dfKang2017.columns = dfKang2017.iloc[0]
dfKang2017.drop(index=0, inplace=True)
dfKang2017.reset_index
dfKang2017["Sample ID"] = dfKang2017["Sample ID"].str.lower()

substitutions = {r"β": "beta"} # fix spelling
dfKang2017.replace(substitutions, regex=True, inplace=True)
dfKang2017.dropna(inplace=True)

dfKang2017
```

```python
# Find common metabolites between STELLA (computationally-obtained) and Kang (experimentally-obtained)
metabolitesKang2017 = list(dfKang2017["Sample ID"]) # Kang2017 metabolites
metabolitesSTELLA = list(dfSTELLA.index)            # STELLA metabolites
commonMetabolites = [m for m in metabolitesKang2017 if m in metabolitesSTELLA]

print("Kang (2017) shows {} metabolites.".format(len(metabolitesKang2017)))
print("STELLA shows {} metabolites.".format(len(metabolitesSTELLA)))
print("There are in total {} common metabolites: {}".format(len(commonMetabolites), commonMetabolites))
```

## The metabolite(s) at stake

### Dataframes preparation

```python
# Keep only common metabolites
dfK = dfKang2017[dfKang2017["Sample ID"].isin(commonMetabolites)]
dfK.set_index("Sample ID", inplace=True)
dfS = dfSTELLA.loc[commonMetabolites]


# Normalize columns
cols = list(dfS.columns)
dfS[cols] = dfS[cols]/dfS[cols].sum()

dfK.T.describe()
```

We want now to split the `dfK` dataset form Kang et al. in two different dataframes, since the columns can be treated as different trials (or experiments, data...) on patients.
This experiments' outputs can be labeled either as Neurotypical/Wild Type (N/WT) or Affected (A/ASD).

```python
# Neurotypical values
dfKN = dfK.filter(regex=r"N\d*$", axis=1)

# Normalize experiments/patients (columns)
cols = list(dfKN.columns)
dfKN[cols] = dfKN[cols]/dfKN[cols].sum()

# benchmark values
bN = dfKN.T.describe().loc[["mean", "std", "min", "max"]]

dfKN.T
```

```python
# ASD values
dfKA = dfK.filter(regex=r"A\d*$", axis=1)

# Normalize experiments/patients (columns)
cols = list(dfKA.columns)
dfKA[cols] = dfKA[cols]/dfKA[cols].sum()

# benchmark values
bA = dfKA.T.describe().loc[["mean", "std", "min", "max"]]

dfKA.T
```

## The average metabolite(s)

```python
dfCounter = dfS.copy()

# Choices: true negative, false positive, true positive, false negative
choicesWT = ["TN", "FP"]
choicesASD = ["TP", "FN"]

for column in dfCounter:
    # Spot true negative and false positive
    conditionsWT = [np.abs(dfCounter[column]-bN.T["mean"]) < np.abs(dfCounter[column]-bA.T["mean"]), # True negative
                    np.abs(dfCounter[column]-bN.T["mean"]) > np.abs(dfCounter[column]-bA.T["mean"])  # False positive 
                   ]
    # Spot true positive and false negative
    conditionsASD = [np.abs(dfCounter[column]-bA.T["mean"]) < np.abs(dfCounter[column]-bN.T["mean"]), # True positive
                     np.abs(dfCounter[column]-bA.T["mean"]) > np.abs(dfCounter[column]-bN.T["mean"])  # False negative
                    ]

    # Neurotypical check
    if "Control" in column:
        dfCounter[column] = np.select(conditionsWT, choicesWT)
    # ASD check
    elif "Autism" in column:
        dfCounter[column] = np.select(conditionsASD, choicesASD)
        
# Count values
TP = dfCounter.isin(["TP"]).sum().sum()
TN = dfCounter.isin(["TN"]).sum().sum()
FP = dfCounter.isin(["FP"]).sum().sum()
FN = dfCounter.isin(["FN"]).sum().sum()

# Compute different F-scores
# Recall Fb(tp, fn, fp, beta=1)
F1 = Fb(TP, FN, FP)
F2 = Fb(TP, FN, FP, 2)
F05 = Fb(TP, FN, FP, 0.5)

print("STELLA's F(beta)-scores:")
print("{:>8} = {}".format("F(1)", F1))
print("{:>8} = {}".format("F(2)", F2))
print("{:>8} = {}".format("F(1/2)", F05))
```

## 1-by-1 comparison for each "experiment" of Kang's dataset

We compare pairwise each STELLA trial with a Kang's experiment dataset, one-by-one with no spefici rule but the order of the experiment (e.g., STELLA$_1$ is compared with Kang$_1$, STELLA$_2$ with Kang$_2$, and so on).

```python
# Create empty dataframe
# Rows: F(1), F(2), F(0.5)
# Cols: number of trials
dfFscores = pd.DataFrame()

# Choices: true negative, false positive, true positive, false negative
choicesWT = ["TN", "FP"]
choicesASD = ["TP", "FN"]

# Reference dataframes
dfKN = dfK.filter(regex=r"N\d*$", axis=1)
dfKA = dfK.filter(regex=r"A\d*$", axis=1)

# Loop over neurotypical values to spot true negative and false positive
for n in dfKN:
    # corresponding ASD column name
    a = "A"+n[1:]

    # Dataframe to analyze
    dfCounter = dfS.copy()

    # For each of STELLA's data spot true negative and false positive / true positive and false negative
    for c in dfCounter:
        # Spot true negative and false positive
        conditionsWT = [np.abs(dfCounter[c]-dfKN[n]) < np.abs(dfCounter[c]-dfKA[a]), # True negative
                        np.abs(dfCounter[c]-dfKN[n]) > np.abs(dfCounter[c]-dfKA[a])  # False positive 
                       ] 
        # Spot true positive and false negative
        conditionsASD = [np.abs(dfCounter[c]-dfKA[a]) < np.abs(dfCounter[c]-dfKN[n]), # True positive
                         np.abs(dfCounter[c]-dfKA[a]) > np.abs(dfCounter[c]-dfKN[n])  # False negative
                        ]

        # Neurotypical check
        if "Control" in column:
            dfCounter[c] = np.select(conditionsWT, choicesWT)
        # ASD check
        elif "Autism" in column:
            dfCounter[c] = np.select(conditionsASD, choicesASD)

    # Count values
    TP = dfCounter.isin(["TP"]).sum().sum()
    TN = dfCounter.isin(["TN"]).sum().sum()
    FP = dfCounter.isin(["FP"]).sum().sum()
    FN = dfCounter.isin(["FN"]).sum().sum()

    # Compute different F-scores
    # Recall Fb(tp, fn, fp, beta=1)
    F1 = Fb(TP, FN, FP)
    F2 = Fb(TP, FN, FP, 2)
    F05 = Fb(TP, FN, FP, 0.5)

    # Build pandas Series
    d = {"F1": F1, "F2": F2, "F05": F05}
    sFscore = pd.Series(data=d, index=["F1", "F2", "F05"])
    sFscore.name = "n"+n[1:]
    #dfFscores.append(sFscore)
    dfFscores = pd.concat([dfFscores, sFscore], axis=1)


dfFscores
```

## Shuffle experiments

Obviously, the previous strategy is not 100% correct, since there may be correlations between the two sets of experiments.
Thus, in order to loose these possible correlations we shall shuffle many times Kang's dataset, and measure for each shuffled dataframe different $F_1$ scores.

In particular, we compare each of STELLA's trials with a couple of Kang's experiments, namely with Neurotypical and ASD data. 
This couple is extracted randomly from the set of possible (N, ASD) couples in `dfK`.
We do this by shuffling the N and ASD columns, respectively, and picking at each iteration step in order the 1st, then the 2nd couple, then the 3rd and so on.

This should give a fairly better computation of the F-score for STELLA's data.

![](pix/shuffle.svg "")

```python
# Create empty dataframe
# Rows: F(1), F(2), F(0.5)
# Cols: number of trials
dfFscores = pd.DataFrame()

# Choices: true negative, false positive, true positive, false negative
choicesWT = ["TN", "FP"]
choicesASD = ["TP", "FN"]

# Reference dataframes
dfKN = dfK.filter(regex=r"N\d*$", axis=1)
dfKA = dfK.filter(regex=r"A\d*$", axis=1)

# Shuffle many times to loose correlations between experiments
nSimulations = 200
for i in range(nSimulations):
    dfn = dfKN.copy().T.sample(frac=1).T
    dfn.columns = list(dfKN.columns)
    dfa = dfKA.copy().T.sample(frac=1).T
    dfa.columns = list(dfKA.columns)

    # Loop over neurotypical values to spot true negative and false positive
    for n in dfn:
        # corresponding ASD column name
        a = "A"+n[1:]

        # Dataframe to analyze
        dfCounter = dfS.copy()

        # For each of STELLA's data spot true negative and false positive / true positive and false negative
        for c in dfCounter:
            # Spot true negative and false positive
            conditionsWT = [np.abs(dfCounter[c]-dfn[n]) < np.abs(dfCounter[c]-dfa[a]), # True negative
                            np.abs(dfCounter[c]-dfn[n]) > np.abs(dfCounter[c]-dfa[a])  # False positive 
                           ] 
            # Spot true positive and false negative
            conditionsASD = [np.abs(dfCounter[c]-dfa[a]) < np.abs(dfCounter[c]-dfn[n]), # True positive
                             np.abs(dfCounter[c]-dfa[a]) > np.abs(dfCounter[c]-dfn[n])  # False negative
                            ]

            # if neurotypical check whether TN or FP
            if "Control" in c:
                dfCounter[c] = np.select(conditionsWT, choicesWT)
            # if ASD check whether TP or FN
            elif "Autism" in c:
                dfCounter[c] = np.select(conditionsASD, choicesASD)

                
        # Count values
        TP = dfCounter.isin(["TP"]).sum().sum()
        TN = dfCounter.isin(["TN"]).sum().sum()
        FP = dfCounter.isin(["FP"]).sum().sum()
        FN = dfCounter.isin(["FN"]).sum().sum()

        # Compute different F-scores
        # Recall Fb(tp, fn, fp, beta=1)
        F1 = Fb(TP, FN, FP)
        F2 = Fb(TP, FN, FP, 2)
        F05 = Fb(TP, FN, FP, 0.5)

        # Build Series
        d = {"F1": F1, "F2": F2, "F05": F05}
        sFscore = pd.Series(data=d, index=["F1", "F2", "F05"])
        sFscore.name = "n"+str((i+1)*int(n[1:])).zfill(4)
        
        # Build DataFrame
        dfFscores = pd.concat([dfFscores, sFscore], axis=1)

dfFscores
```

We shall compare now STELLA's $F_1$-score to the scores computed in [Yin et al., 2020](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC7746778/), Figure 3.
Yin et al. compute wheter a specific metabolite is present or absent in each patient.
It is thus a score associated with the ability of the method to spot the presence of key metabolites.
On the other hand, our method computes the abundancy of the metabolites in neurotypical/autistic individuals.
Thus, it is not that correct to compare the two $F_1$ scores, since they are associated with different measurements.

However, STELLA's $F_1$ score is in general larger to those of the other computational methods, meaning that in general our method is not that inaccurate as the others, as reported by Yin et al.

### Sensitivity, Specificity, Precision and Recall

We shall compute now some statistical parameters (namely sensitivity, specificity, precision and recall) starting from the obtained true/false positive/negative values.
In this way it is possible to plot a confusion map about the relations between these data.
Moreover, in this way we have another element through which we can compare STELLA with the other computational methods (e.g., MIMOSA2, PelonPamm, MangoSteen, ...).

The idea is to reproduce something similar to Fig.4 of [Noekcker et al., 2022](https://academic.oup.com/bioinformatics/article/38/6/1615/6499258).

- sensitivity, recall, hit rate, or true positive rate (TPR)
    $$
      \text{TPR} = \frac{\text{TP}}{\text{P}} = \frac{\text{TP}}{\text{TP}+\text{FN}}
    $$
- specificity, selectivity or true negative rate (TNR)
    $$
      \text{TNR} = \frac{\text{TN}}{\text{N}} = \frac{\text{TN}}{\text{TN}+\text{FP}}
    $$
- precision or positive predictive value (PPV)
    $$
      \text{PPV} = \frac{\text{TP}}{\text{TP}+\text{FP}}
    $$

We report here STELLA's [confusion map](https://www.wikiwand.com/en/Confusion_matrix) for Kang's dataset.

```python
# Normalization
tot = (TN+FP+TP+FN)

print("Data report for {} shuffle trials".format(nSimulations))
print("True negative values:  {}".format(TN/tot))
print("True positive values:  {}\n".format(TP/tot))
print("False positive values: {}".format(FP/tot))
print("False negative values: {}\n".format(FN/tot))

# Confusion matrix
confMatrix = np.array([[TN/tot, TP/tot], [FN/tot, FP/tot]])

# Plot
ax = sns.heatmap(confMatrix, annot=True)
ax.set_xticklabels(["Negative", "Positive"]);
ax.set_yticklabels(["True", "False"]);

"""
------------
[200 simulations]
  True negative values:  0.28674698795180725
  True positive values:  0.21686746987951808
  False positive values: 0.19518072289156627
  False negative values: 0.30120481927710846
-------------
[500 simulations]
  True negative values:  0.1927710843373494
  True positive values:  0.3108433734939759
  False positive values: 0.2891566265060241
  False negative values: 0.20722891566265061
------------
[200 simulations]
  True negative values:  0.2
  True positive values:  0.30120481927710846
  False positive values: 0.2819277108433735
  False negative values: 0.21686746987951808
""";
```
