# $Z$-score

We want to compute the $Z$-score for the ASD experimental data in order to find out if a specific metabolite tends to be consumed or produced.
Then, we shall compare STELLA's output with the neurotypical- and ASD-related associated scores.

Wherever there is a match we assign a TN or TP value, respectively. 
In case of mismatches between the scores, false values are assigned.

```python
import html

import matplotlib.pyplot as plt
import matplotlib.colors # https://matplotlib.org/2.0.2/users/colormaps.html
import numpy as np
import pandas as pd
import seaborn as sns

from glob import glob

# Be sure the code is correct when excluding warnings!
import warnings
warnings.filterwarnings("ignore") # ignore boring warnings
```

```python
def Fb(tp, fn, fp, beta=1):
    """Compute F_beta score for a given number of true/false positive/negative values
    Parameters
    ----------
    tp: number of true positive values
    fn: number of false negative values
    fp: number of false positive values
    beta: value for which recall is considered beta times as important as precision:
          Default value is beta=1
    """
    return ((1+beta*beta)*tp) / ((1+beta*beta)*tp + beta*beta*fn + fp)
```

```python
# Import whole STELLA data
dfSTELLA = pd.read_csv("asd/patient-metabolite-matrix.csv")
dfSTELLA.set_index("Unnamed: 0", inplace=True)

dfSTELLA
```

```python
# Z-score dataframe for neurotypical values
# z = (value-mean)/(standard deviation)

dfKN = dfK.filter(regex=r"N\d*$", axis=1)
dfKN = ((dfKN.T - dfKN.T.mean())/(dfKN.T.std())).T
dfKN.dropna(inplace=True)
zN = dfKN.T.describe().loc["mean"].T
zN

"""
ax = sns.heatmap(dfKN, cmap="viridis")
ax.set_ylabel("");
ax.set_xlabel("");
ax.set_title(r"Kang (2017) Neurotypical $z$-score");
plt.savefig("plots/kang2017-zscore-N.pdf", bbox_inches="tight")
""";
```

```python
# Z-score dataframe for neurotypical values
# z = (value-mean)/(standard deviation)

dfKA = dfK.filter(regex=r"A\d*$", axis=1)
dfKA = ((dfKA.T - dfKA.T.mean())/(dfKA.T.std())).T
dfKA.dropna(inplace=True)
zA = dfKA.T.describe().loc["mean"].T
zA

"""
ax = sns.heatmap(dfKA, cmap="viridis")
ax.set_ylabel("");
ax.set_xlabel("");
ax.set_title(r"Kang (2017) ASD $z$-score");
#plt.savefig("plots/kang2017-zscore-ASD.pdf", bbox_inches="tight")
""";
```

```python
# Put FP, FN, TP, TN similarly to 7th script
# if column is Control, compare with zN
# if columns is ASD, compare with zA

dfS = dfSTELLA.loc[commonMetabolites]

# For each of STELLA's data spot true negative and false positive / true positive and false negative
for s in dfS:
    # if neurotypical check whether TN (same sign) or FP (different sign)
    if "Control" in s:
        dfS[s] *= zN
        dfS[s] = np.where(dfS[s]>=0, "TN", "FP")
        
    # elif ASD check whether TP (same sign) or FN (different sign)
    elif "Autism" in s:
        dfS[s] *= zN
        dfS[s] = np.where(dfS[s]>=0, "TP", "FN")

# Count values
TP = dfS.isin(["TP"]).sum().sum()
TN = dfS.isin(["TN"]).sum().sum()
FP = dfS.isin(["FP"]).sum().sum()
FN = dfS.isin(["FN"]).sum().sum()

# Compute different F-scores
# Recall Fb(tp, fn, fp, beta=1)
F1 = Fb(TP, FN, FP)
F2 = Fb(TP, FN, FP, 2)
F05 = Fb(TP, FN, FP, 0.5)

print("F1 score: {}".format(F1))
print("F2 score: {}".format(F2))
print("F05 score: {}".format(F05))
```

```python
# Normalization
tot = (TN+FP+TP+FN)

# Statistical quantities of interest
sensitivity = TP/(TP+FN)
specificity = TN/(TN+FP)
precision = TP/(TP+FP)

print("True negative values:  {}".format(TN/tot))
print("True positive values:  {}".format(TP/tot))
print("False positive values: {}".format(FP/tot))
print("False negative values: {}\n".format(FN/tot))
print("Sensitivity = {}".format(sensitivity))
print("Specificity = {}".format(specificity))
print("  Precision = {}\n".format(precision))

# Confusion matrix
confMatrix = np.array([[TN/tot, TP/tot], [FN/tot, FP/tot]])

# Plot
ax = sns.heatmap(confMatrix, annot=True)
ax.set_xticklabels(["Negative", "Positive"]);
ax.set_yticklabels(["True", "False"]);
ax.set_title("Confusion Map");

plt.savefig("plots/confusion-map-totNorm.pdf", bbox_inches="tight")
```

Unfortunately, it makes sense that the abundance of TP, TN, FP and FN are very similar.
In fact, one may observe how the experimental data (Fig.1) from Kang do not display major differences between healthy and ASD patients.

Being the data uniformly distributed among the two patient categories there is no marked difference between the two, while one may have expected a 2-block heatmap.
Similarly, STELLA's data appears uniformly distributed too among all patients, with the exception of one only ASD patient (the 42nd).

Thus, statistically for each of the two blocks half of the data will be well-set (i.e., TP or TN), and the remaining half will be wrongly predicted (that is, FN or FP).

## OTU-metabolite correlation

WE WANT TO CHECK IF THERE IS SOME SORT OF CORRELATION BETWEEN METABOLITES AND OTUs, AND TO QUANTIFY IT

We try to compute just a single $z$-score, independently of the patient health status.

- Quando è prodotto per STELLA e per z -> TP
- Quando è prodotto per STELLA e non per z -> FP
- Quando non è prodotto per STELLA e non lo è per z -> TN
- Quando non è prodotto per STELLA e lo è per z -> FN


STELLA = $\pm 2$, Kang = $\pm 1$ $\implies$ summing the two dataframes there are four possible outputs, that is $\pm 1$ and $\pm 3$, that can be associated with FP/FN and TP/TN, respectively.

```python
dfz = dfK.copy()
dfz = ((dfz.T - dfz.T.mean())/(dfz.T.std())).T
dfz.dropna(inplace=True)
dfz
sz = dfz.T.describe().loc["mean"].T
sz = sz/np.abs(sz)
sz
```

```python
# Dataframe to analyze
dfS = dfSTELLA.loc[commonMetabolites]
dfCounter = dfS.copy()
choicesWT = ["TN", "FP"]
choicesASD = ["TP", "FN"]

# We consider null metabolite as produced
dfCounter[dfCounter == 0] = 1
dfCounter = dfCounter/np.abs(dfCounter)*2

# For each of STELLA's data spot true negative and false positive / true positive and false negative
for c in dfCounter:
    dfCounter[c] += sz
    
    # Spot true negative and false positive
    conditionsWT = [dfCounter[c] == -3, # TN
                    dfCounter[c] == +1  # FP
                   ] 
    # Spot true positive and false negative
    conditionsASD = [dfCounter[c] == +3, # TP
                     dfCounter[c] == -1  # FN
                    ] 

    # if neurotypical check whether TN or FP
    if "Control" in c:
        dfCounter[c] = np.select(conditionsWT, choicesWT)
    # if ASD check whether TP or FN
    elif "Autism" in c:
        dfCounter[c] = np.select(conditionsASD, choicesASD)


# Count values
TP = dfCounter.isin(["TP"]).sum().sum()
TN = dfCounter.isin(["TN"]).sum().sum()
FP = dfCounter.isin(["FP"]).sum().sum()
FN = dfCounter.isin(["FN"]).sum().sum()

# Compute different F-scores
# Recall Fb(tp, fn, fp, beta=1)
F1 = Fb(TP, FN, FP)
F2 = Fb(TP, FN, FP, 2)
F05 = Fb(TP, FN, FP, 0.5)
print(F1, F2, F05)
```

```python
# Normalization
tot = (TN+FP+TP+FN)

# Statistical quantities of interest
sensitivity = TP/(TP+FN)
specificity = TN/(TN+FP)
precision = TP/(TP+FP)

print("True negative values:  {}".format(TN/tot))
print("True positive values:  {}".format(TP/tot))
print("False positive values: {}".format(FP/tot))
print("False negative values: {}\n".format(FN/tot))
print("Sensitivity = {}".format(sensitivity))
print("Specificity = {}".format(specificity))
print("  Precision = {}\n".format(precision))

# Confusion matrix
confMatrix = np.array([[TN/(TN+FN), TP/(TP+FP)], [FN/(TN+FN), FP/(TP+FP)]])
confMatrixTot = np.array([[TN/(tot), TP/(tot)], [FN/(tot), FP/(tot)]])

# Plot
ax = sns.heatmap(confMatrix, annot=True)
ax.set_xticklabels(["Negative", "Positive"]);
ax.set_yticklabels(["True", "False"]);
ax.set_title("Confusion Map");

plt.savefig("plots/confusion-map-colsNorm.pdf", bbox_inches="tight")
```
